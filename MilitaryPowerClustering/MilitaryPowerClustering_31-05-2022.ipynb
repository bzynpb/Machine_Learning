# Modelling

## Hopkins Test 
#### Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express no clusterability, a score tending to 1 express an uniform distrubution so clustering can't be useful and  0 express  not an uniform distrubution hence clusterering can be useful.

```python
from pyclustertend import hopkins # 1 express an uniform distrubution so clustering can't be useful
```

```python
hopkins(X, X.shape[0])
```

## Clustering with K-means

### Elbow Method 
### We use this method to find optimal k value by looking up to break point of graph. 

```python
ssd = [] # sum of squared distance

K = range(2,10)

for k in K:
    model = KMeans(n_clusters=k)
    model.fit(X)
    ssd.append(model.inertia_)
# INERTIA, noktalarin kendi icinde sÄ±kÄ± sÄ±kÄ±ya kÃ¼melenmeyi gÃ¶steriyor, 
# inertia degerinin kÃ¼Ã§Ã¼k olmasÄ± istenir,merkez(centroid) etrafÄ±nda kÃ¼melenme.
# DÃ¼ÅÃ¼k olmasÄ± kÃ¼me iÃ§i iyi kÃ¼melenmeyi(yoÄunlaÅmayÄ±) gÃ¶sterir.
```

```python
ssd
# sum of squared distance 
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method")
# degisen k degerine karsilik inertia degeri 
# keskin dususun durdugu ilk nokta, optimal k noktasi
#Â 2 ve 3 keskin dusus diyebiliriz 
```

```python
pd.Series(ssd).diff()
```

```python
df1_diff = pd.DataFrame(-pd.Series(ssd).diff()).rename(index=lambda x: x + 1)
df1_diff
# keskin dususu anlamak icin
```

```python
df1_diff.plot(kind='bar');
```

```python
kmeans = KMeans()
visu = KElbowVisualizer(kmeans, k = (2,10))
visu.fit(X)
visu.poof();
# optimal k degeri 4 oldugunu gosteriyor 
```

### Silhouette analysis

```python
from sklearn.metrics import silhouette_score
# modelimizin baÅarÄ±sÄ± Ã¶lÃ§Ã¼yoruz, hem yoÄun gruplaÅma  hem de gruplar arasÄ± uzaklÄ±ÄÄ± Ã¶lÃ§erek
# kÃ¼melemenin baÅarÄ±lÄ± mi deÄil mi?
# 1 e ne kadar yakÄ±nsa o kadar iyi
```

```python
silhouette_score(X, kmeans.labels_)
# ne kadar yuksek olursa o kadar iyi kumelenme. var demektir 
```

```python
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=101)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
## hem optimal k degerine bakmali (initaer kucuk olmali)
## siluet square degerinin buyuk olmasi 
## 3 siluete yuksek, inertia de yuksek 
## optimal deger 4 cunku yukarida inertia degeri dusuk
```

## Hierarchical Clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

### Dendogram

```python
hc_ward = linkage(y=X, method="ward")
hc_complete = linkage(X, "complete")
hc_average = linkage(X, "average")
hc_single = linkage(X, "single")
```

```python
hc_ward = linkage(y=X, method="ward")  # eklendiÄi sÄ±nÄ±fÄ±n varyansÄ±nÄ± dÃ¼ÅÃ¼recek Åekile ata
hc_complete = linkage(X, "complete")  # en uzak noktaya gÃ¶re en kÃ¼Ã§Ã¼k deÄer
hc_average = linkage(X, "average")  # ortalama
hc_single = linkage(X, "single")  #en yakÄ±n noktaya gÃ¶re en kÃ¼Ã§Ã¼k deÄer
```

```python
plt.figure(figsize=(20, 12))

plt.subplot(221)
plt.title("Ward")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_ward, leaf_font_size=10, truncate_mode='lastp', p=10)

plt.subplot(222)
plt.title("Complete")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_complete, leaf_font_size=10, truncate_mode='lastp', p=10)

plt.subplot(223)
plt.title("Average")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_average, leaf_font_size=10, truncate_mode='lastp', p=10)

plt.subplot(224)
plt.title("Single")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_single, leaf_font_size=10, truncate_mode='lastp', p=10)
```
